```python
from model_verification import ModelSelection
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C, WhiteKernel as Wh


if __name__ == "__main__":
    path = "../../../data/processed_data/master_data_width.csv"
    kernel = C(1.0, (1e-2,1e2)) * RBF(1.0, (1e-2,1e2)) + Wh(0.01, (1e-2,1e2))
    ms = ModelSelection(path)
    ms.standardization()
    ms.gaussian_regression(kernel)
    result = ms.cross_validation()
    print("RMSE: ", result)
    ms.visualization(method="GPR")
```
